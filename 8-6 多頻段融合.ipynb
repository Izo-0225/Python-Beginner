```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

apple = cv2.imread('d:/ipython/blend/apple.jpg',3)
orange = cv2.imread('d:/ipython/blend/orange_resized.jpg',3)
mask= cv2.imread('d:/ipython/blend/mask_resized.jpg',0).astype(np.float32)
```

```python
# to show mask in gray. 
mask_gray= cv2.imread('d:/ipython/blend/mask_resized.jpg',1)
plt.imshow(mask_gray)
```

```python
plt.imshow(cv2.cvtColor(apple, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(orange, cv2.COLOR_BGR2RGB))
```

```python
a_B, a_G, a_R = cv2.split(orange)
b_B, b_G, b_R = cv2.split(apple)

b = (a_B * (mask/255.0)) + (b_B * (1.0 - (mask/255.0)))
g = (a_G * (mask/255.0)) + (b_G * (1.0 - (mask/255.0)))
r = (a_R * (mask/255.0)) + (b_R * (1.0 - (mask/255.0)))
output = cv2.merge((b,g,r))
```

```python
cv2.imwrite("d:/ipython/blend/fruit.jpg", output) 

img=cv2.imread('d:/ipython/blend/fruit.jpg', 3)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
