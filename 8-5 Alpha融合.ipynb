```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("d:/ipython/blend/Trump_resized.jpg")
img2 = cv2.imread('d:/ipython/blend/Biden 1.jpg')
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
```

```python
dist = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
cv2.imshow("Dist", dist)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dist1 = cv2.addWeighted(img1, 0.3, img2, 0.7, 0)
cv2.imshow("Dist1", dist1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
